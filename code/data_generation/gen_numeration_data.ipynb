filename = 'numeration_word2num.json'

structure of data is a dictionary of dictionaries
```
{
    'positive int': {'d1': [] , 'd2': [],.. 'd7': []},
    'negative int': {'d1': [] , 'd2': [],.. 'd7': []},
    'positive frac': {'d1': [] , 'd2': [],.. 'd7': []},
    'negative frac': {'d1': [] , 'd2': [],.. 'd7': []},
}
```

d1 represents 1 digit numbers, each entry in list [] is a string of 1 digit number for it.

fractions are all rounded off to 2 decimal places.

Number of examples are as follows:

|datatype|d1|d2|d3|d4|d5|d6|d7|Total|
|---|---|---|---|---|---|---|---|---|
|positive int|10|90|200|200|200|200|200|1100|
|negative int|10|90|200|200|200|200|200|1100|
|positive frac|200|200|200|200|200|200|200|1400|
|negative frac|200|200|200|200|200|200|200|1400|

total examples = 5000

# Numeration word2num


````python
import json
import random
from num2words import num2words
````

````python
NUM_PER_CATEGORY = 200
````

````python
# generate positive integers of 1,2,3,4,5,6,7 digits
positive_int_d1 =[]
positive_int_d2 =[]
positive_int_d3 =[]
positive_int_d4 =[]
positive_int_d5 =[]
positive_int_d6 =[]
positive_int_d7 =[]

# generate all numbers till 999
for i in range(100):
    if i < 10:
        positive_int_d1.append(num2words(i))
    else:
        positive_int_d2.append(num2words(i))

random.seed(21)
# for each digit generate random numbers
for digit in  [3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set() # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY: 
        # generate random number
        num = random.randint(start,end)
        # check if number is already generated
        if num in gen_num:
            continue
        # add number to set
        gen_num.add(num)
        i += 1
        if digit == 3:
            positive_int_d3.append(num2words(num))
        elif digit == 4:
            positive_int_d4.append(num2words(num))
        elif digit == 5:
            positive_int_d5.append(num2words(num))
        elif digit == 6:
            positive_int_d6.append(num2words(num))
        elif digit == 7:
            positive_int_d7.append(num2words(num))
````

````python
# generate negative integers of 1,2,3,4,5,6,7 digits
negative_int_d1 =[]
negative_int_d2 =[]
negative_int_d3 =[]
negative_int_d4 =[]
negative_int_d5 =[]
negative_int_d6 =[]
negative_int_d7 =[]

# generate all numbers till 999
for i in range(100):
    if i < 10:
        negative_int_d1.append(num2words(-1*i))
    else:
        negative_int_d2.append(num2words(-1*i))

random.seed(42)
# for each digit generate random numbers
for digit in  [3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set() # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY: 
        # generate random number
        num = -1* random.randint(start,end)
        # check if number is already generated
        if num in gen_num:
            continue
        gen_num.add(num)
        i += 1
        if digit == 3:
            negative_int_d3.append(num2words(num))
        elif digit == 4:
            negative_int_d4.append(num2words(num))
        elif digit == 5:
            negative_int_d5.append(num2words(num))
        elif digit == 6:
            negative_int_d6.append(num2words(num))
        elif digit == 7:
            negative_int_d7.append(num2words(num))
````

````python
# generate positive fractions of 1,2,3,4,5,6,7 digits all are rounded to 2 decimal places
positive_frac_d1 =[]
positive_frac_d2 =[]
positive_frac_d3 =[]
positive_frac_d4 =[]
positive_frac_d5 =[]
positive_frac_d6 =[]
positive_frac_d7 =[]

random.seed(63)
for digit in [1,2,3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set() # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY:
        # generate random number 
        num = round(random.uniform(start,end),2)
        # check if number is already generated
        if num in gen_num:
            continue
        gen_num.add(num)
        i += 1
        if digit == 1:
            positive_frac_d1.append(num2words(num))
        elif digit == 2:
            positive_frac_d2.append(num2words(num))
        elif digit == 3:
            positive_frac_d3.append(num2words(num))
        elif digit == 4:
            positive_frac_d4.append(num2words(num))
        elif digit == 5:
            positive_frac_d5.append(num2words(num))
        elif digit == 6:
            positive_frac_d6.append(num2words(num))
        elif digit == 7:
            positive_frac_d7.append(num2words(num))

````

````python
# generate negative fractions of 1,2,3,4,5,6,7 digits all are rounded to 2 decimal places
negative_frac_d1 =[]
negative_frac_d2 =[]
negative_frac_d3 =[]
negative_frac_d4 =[]
negative_frac_d5 =[]
negative_frac_d6 =[]
negative_frac_d7 =[]

random.seed(84)
for digit in [1,2,3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set()    # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY: 
        # generate random number
        num = -1* round(random.uniform(start,end),2)
        if num in gen_num:
            continue
        gen_num.add(num)
        i += 1
        if digit == 1:
            negative_frac_d1.append(num2words(num))
        elif digit == 2:
            negative_frac_d2.append(num2words(num))
        elif digit == 3:
            negative_frac_d3.append(num2words(num))
        elif digit == 4:
            negative_frac_d4.append(num2words(num))
        elif digit == 5:
            negative_frac_d5.append(num2words(num))
        elif digit == 6:
            negative_frac_d6.append(num2words(num))
        elif digit == 7:
            negative_frac_d7.append(num2words(num))
````

````python
data = {'positive int': {'d1': positive_int_d1, 'd2': positive_int_d2, 'd3': positive_int_d3, 'd4': positive_int_d4, 'd5': positive_int_d5, 'd6': positive_int_d6, 'd7': positive_int_d7},
        'negative int': {'d1': negative_int_d1, 'd2': negative_int_d2, 'd3': negative_int_d3, 'd4': negative_int_d4, 'd5': negative_int_d5, 'd6': negative_int_d6, 'd7': negative_int_d7},
        'positive frac': {'d1': positive_frac_d1, 'd2': positive_frac_d2, 'd3': positive_frac_d3, 'd4': positive_frac_d4, 'd5': positive_frac_d5, 'd6': positive_frac_d6, 'd7': positive_frac_d7},
        'negative frac': {'d1': negative_frac_d1, 'd2': negative_frac_d2, 'd3': negative_frac_d3, 'd4': negative_frac_d4, 'd5': negative_frac_d5, 'd6': negative_frac_d6, 'd7': negative_frac_d7}
        }
````

````python
with open('../../data/numeration_word2num.json', 'w') as f:
    json.dump(data, f, indent=4)
````
