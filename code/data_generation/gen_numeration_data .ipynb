# Numeration word2num


````python
import json
import random
from num2words import num2words
````

````python
NUM_PER_CATEGORY = 200
````

````python
# generate positive integers of 1,2,3,4,5,6,7 digits
positive_int_d1 =[]
positive_int_d2 =[]
positive_int_d3 =[]
positive_int_d4 =[]
positive_int_d5 =[]
positive_int_d6 =[]
positive_int_d7 =[]

# generate all numbers till 999
for i in range(100):
    if i < 10:
        positive_int_d1.append(num2words(i))
    else:
        positive_int_d2.append(num2words(i))

random.seed(21)
# for each digit generate random numbers
for digit in  [3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set() # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY: 
        # generate random number
        num = random.randint(start,end)
        # check if number is already generated
        if num in gen_num:
            continue
        # add number to set
        gen_num.add(num)
        i += 1
        if digit == 3:
            positive_int_d3.append(num2words(num))
        elif digit == 4:
            positive_int_d4.append(num2words(num))
        elif digit == 5:
            positive_int_d5.append(num2words(num))
        elif digit == 6:
            positive_int_d6.append(num2words(num))
        elif digit == 7:
            positive_int_d7.append(num2words(num))
````

````python
# generate negative integers of 1,2,3,4,5,6,7 digits
negative_int_d1 =[]
negative_int_d2 =[]
negative_int_d3 =[]
negative_int_d4 =[]
negative_int_d5 =[]
negative_int_d6 =[]
negative_int_d7 =[]

# generate all numbers till 999
for i in range(100):
    if i < 10:
        negative_int_d1.append(num2words(-1*i))
    else:
        negative_int_d2.append(num2words(-1*i))

random.seed(42)
# for each digit generate random numbers
for digit in  [3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set() # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY: 
        # generate random number
        num = -1* random.randint(start,end)
        # check if number is already generated
        if num in gen_num:
            continue
        gen_num.add(num)
        i += 1
        if digit == 3:
            negative_int_d3.append(num2words(num))
        elif digit == 4:
            negative_int_d4.append(num2words(num))
        elif digit == 5:
            negative_int_d5.append(num2words(num))
        elif digit == 6:
            negative_int_d6.append(num2words(num))
        elif digit == 7:
            negative_int_d7.append(num2words(num))
````

````python
# generate positive decimals of 1,2,3,4,5,6,7 digits all are rounded to 2 decimal places
positive_dec_d1 =[]
positive_dec_d2 =[]
positive_dec_d3 =[]
positive_dec_d4 =[]
positive_dec_d5 =[]
positive_dec_d6 =[]
positive_dec_d7 =[]

random.seed(63)
for digit in [1,2,3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set() # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY:
        # generate random number 
        num = round(random.uniform(start,end),2)
        # check if number is already generated
        if num in gen_num:
            continue
        gen_num.add(num)
        i += 1
        if digit == 1:
            positive_dec_d1.append(num2words(num))
        elif digit == 2:
            positive_dec_d2.append(num2words(num))
        elif digit == 3:
            positive_dec_d3.append(num2words(num))
        elif digit == 4:
            positive_dec_d4.append(num2words(num))
        elif digit == 5:
            positive_dec_d5.append(num2words(num))
        elif digit == 6:
            positive_dec_d6.append(num2words(num))
        elif digit == 7:
            positive_dec_d7.append(num2words(num))

````

````python
# generate negative decimals of 1,2,3,4,5,6,7 digits all are rounded to 2 decimal places
negative_dec_d1 =[]
negative_dec_d2 =[]
negative_dec_d3 =[]
negative_dec_d4 =[]
negative_dec_d5 =[]
negative_dec_d6 =[]
negative_dec_d7 =[]

random.seed(84)
for digit in [1,2,3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_num = set()    # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY: 
        # generate random number
        num = -1* round(random.uniform(start,end),2)
        if num in gen_num:
            continue
        gen_num.add(num)
        i += 1
        if digit == 1:
            negative_dec_d1.append(num2words(num))
        elif digit == 2:
            negative_dec_d2.append(num2words(num))
        elif digit == 3:
            negative_dec_d3.append(num2words(num))
        elif digit == 4:
            negative_dec_d4.append(num2words(num))
        elif digit == 5:
            negative_dec_d5.append(num2words(num))
        elif digit == 6:
            negative_dec_d6.append(num2words(num))
        elif digit == 7:
            negative_dec_d7.append(num2words(num))
````

````python
data = {'positive int': {'d1': positive_int_d1, 'd2': positive_int_d2, 'd3': positive_int_d3, 'd4': positive_int_d4, 'd5': positive_int_d5, 'd6': positive_int_d6, 'd7': positive_int_d7},
        'negative int': {'d1': negative_int_d1, 'd2': negative_int_d2, 'd3': negative_int_d3, 'd4': negative_int_d4, 'd5': negative_int_d5, 'd6': negative_int_d6, 'd7': negative_int_d7},
        'positive dec': {'d1': positive_dec_d1, 'd2': positive_dec_d2, 'd3': positive_dec_d3, 'd4': positive_dec_d4, 'd5': positive_dec_d5, 'd6': positive_dec_d6, 'd7': positive_dec_d7},
        'negative dec': {'d1': negative_dec_d1, 'd2': negative_dec_d2, 'd3': negative_dec_d3, 'd4': negative_dec_d4, 'd5': negative_dec_d5, 'd6': negative_dec_d6, 'd7': negative_dec_d7}
        }
````

# Numeration Integer addition between number in digit and word form

filename = 'numeration_int_addn.json'

structure of data is a dictionary of dictionaries 

dw: means one of numbers in a,b is in word form and other in digit form

ww: means both numbers are in word form
```
{
    'both positive':{   
                        'dw': {'d1': [] , 'd2': [],.. 'd7': []},
                        'ww': {'d1': [] , 'd2': [],.. 'd7': []}
                    },
    'word neg digit pos':{   
                        'dw': {'d1': [] , 'd2': [],.. 'd7': []},
                        'ww': {'d1': [] , 'd2': [],.. 'd7': []}
                    },
    'both negative':{   
                        'dw': {'d1': [] , 'd2': [],.. 'd7': []},
                        'ww': {'d1': [] , 'd2': [],.. 'd7': []}
                    },
}
```

d1 represents 1 digit numbers, each entry in list [] is a [a,b,sum_word,sum_digit]

where a,b are numbers (in word or digit form) and sum_word is sum of a and b in word form and sum_digit is sum of a and b in digit form. to avoid bias due to position, in dw setting, half of the examples have a as digit and b as word and other half have a as word and b as digit.

Number of examples are as follows:

|category|datatype|d1|d2|d3|d4|d5|d6|d7|Total|
|---|---|---|---|---|---|---|---|---|---|
|both positive|dw|100|150|150|150|150|150|150|1000|
|both positive|ww|100|150|150|150|150|150|150|1000|
|---|---|---|---|---|---|---|---|---|---|
|word neg digit pos|dw|90|150|150|150|150|150|150|990|
|word neg digit pos|ww|90|150|150|150|150|150|150|990|
|---|---|---|---|---|---|---|---|---|---|
|both negative|dw|81|150|150|150|150|150|150|981|
|both negative|ww|81|150|150|150|150|150|150|981|


total examples = 5942

Note: zero is counted as only positive integer and not negative integer

````python
import json
import random
from num2words import num2words
````

````python
data = {'both positive': {}, 'word neg digit pos' : {}, 'both negative' : {}}
NUM_PER_CATEGORY = 150
````

````python
# first create dataset for both positive
d1_dw,d1_ww=[],[]
d2_dw,d2_ww=[],[]
d3_dw,d3_ww=[],[]
d4_dw,d4_ww=[],[]
d5_dw,d5_ww=[],[]
d6_dw,d6_ww=[],[]
d7_dw,d7_ww=[],[]


# for d1, generate all possible combinations of 2 numbers and their sum
for num1 in range(10):
    for num2 in range(10):
        # get the sum
        sum = num1 + num2

        # create the example list
        curr_example_dw = []
        # decide the index of number in digit form
        index_of_digit = random.randint(0,1)
        if index_of_digit == 0:
            curr_example_dw.append(num1)
            curr_example_dw.append(num2words(num2))
        else:
            curr_example_dw.append(num2words(num1))
            curr_example_dw.append(num2)
        # add the sum in word form
        curr_example_dw.append(num2words(sum))
        # add the sum in digit form
        curr_example_dw.append(sum)
        # add the example to the dataset
        d1_dw.append(curr_example_dw)

        # create the example list for word form
        curr_example_ww = []
        curr_example_ww.append(num2words(num1))
        curr_example_ww.append(num2words(num2))
        curr_example_ww.append(num2words(sum))
        curr_example_ww.append(sum)
        d1_ww.append(curr_example_ww)
    
# from d2 onwards generate random 150 examples
random.seed(21)
for digit in [2,3,4,5,6,7]:
    start = 10**(digit-1)
    end = 10**digit
    gen_set = set() # to avoid duplicates
    i = 0
    while i<NUM_PER_CATEGORY:
        num1 = random.randint(start,end)
        num2 = random.randint(start,end)
        sum = num1 + num2
        # check if number is already generated
        if (num1,num2) in gen_set:
            continue
        gen_set.add((num1,num2))
        i += 1


        # create the example list
        curr_example_dw = []
        # decide the index of number in digit form
        index_of_digit = random.randint(0,1)
        if index_of_digit == 0:
            curr_example_dw.append(num1)
            curr_example_dw.append(num2words(num2))
        else:
            curr_example_dw.append(num2words(num1))
            curr_example_dw.append(num2)
        # add the sum in word form
        curr_example_dw.append(num2words(sum))
        # add the sum in digit form
        curr_example_dw.append(sum)
        # add the example to the dataset

        curr_example_ww = []
        curr_example_ww.append(num2words(num1))
        curr_example_ww.append(num2words(num2))
        curr_example_ww.append(num2words(sum))
        curr_example_ww.append(sum)

        if digit == 2:
            d2_dw.append(curr_example_dw)
            d2_ww.append(curr_example_ww)
        elif digit == 3:
            d3_dw.append(curr_example_dw)
            d3_ww.append(curr_example_ww)
        elif digit == 4:
            d4_dw.append(curr_example_dw)
            d4_ww.append(curr_example_ww)
        elif digit == 5:
            d5_dw.append(curr_example_dw)
            d5_ww.append(curr_example_ww)
        elif digit == 6:
            d6_dw.append(curr_example_dw)
            d6_ww.append(curr_example_ww)
        elif digit == 7:
            d7_dw.append(curr_example_dw)
            d7_ww.append(curr_example_ww)

# add in final data
data['both positive']['dw']= {'d1': d1_dw, 'd2': d2_dw, 'd3': d3_dw, 'd4': d4_dw, 'd5': d5_dw, 'd6': d6_dw, 'd7': d7_dw}
data['both positive']['ww']= {'d1': d1_ww, 'd2': d2_ww, 'd3': d3_ww, 'd4': d4_ww, 'd5': d5_ww, 'd6': d6_ww, 'd7': d7_ww}
    
````
