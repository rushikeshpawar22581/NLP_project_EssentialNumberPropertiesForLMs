filename = 'diff_order_int_addition.json'

structure of data is a dictionary of dictionaries
```
{
    'both pos':                 {'d41': [], 'd42': [], 'd43': [],.. 'd101': [], 'd102': [], 'd103': []},
    'bignum pos smallnum neg':  {'d41': [], 'd42': [], 'd43': [],.. 'd101': [], 'd102': [], 'd103': []},
    'bignum neg smallnum pos':  {'d41': [], 'd42': [], 'd43': [],.. 'd101': [], 'd102': [], 'd103': []},
    'both neg':                 {'d41': [], 'd42': [], 'd43': [],.. 'd101': [], 'd102': [], 'd103': []},
}
```

d41 represents bigger number has 4 digits and smaller number has 1 digit numbers, each entry in list [] is a list of [a,b,c] such that 50% are [big_num,small_num,their_sum] and 50% are [small_num,big_num,their_sum]

````python
import json
import random
NUM_PER_CATEGORY = 100
````

````python
# generate positive integers
both_pos_dict = {}
for i in range(4,11):
    for j in range(1,4):
        both_pos_dict["d" + str(i) + str(j)] = []

random.seed(21)
# for each digit generate random numbers
for big_num_digit in  [4,5,6,7,8,9,10]:
    big_num_start = 10**(big_num_digit-1)
    big_num_end = 10**big_num_digit-1 

    for small_num_digit in [1,2,3]:
        small_num_start = 10**(small_num_digit-1)
        small_num_end = 10**small_num_digit-1

        gen_num = set() # to avoid duplicates
        i = 0
        while i<NUM_PER_CATEGORY: 
            # generate random number
            bignum = random.randint(big_num_start, big_num_end)
            smallnum = random.randint(small_num_start, small_num_end)
            # check if number is already generated
            if (bignum,smallnum) in gen_num:
                continue
            i += 1
            # add number to set
            gen_num.add((bignum,smallnum))
            sum = bignum + smallnum

            # to decide which number at which index
            random_num = random.randint(0,1)
            curr_ex = []
            if random_num == 0:
                curr_ex.append(bignum)
                curr_ex.append(smallnum)
            else:
                curr_ex.append(smallnum)
                curr_ex.append(bignum)
            curr_ex.append(sum)

            list_name = 'd'+str(big_num_digit) + str(small_num_digit)
            both_pos_dict[list_name].append(curr_ex)
            
````
